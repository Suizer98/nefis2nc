Streamline Quiver Plot
--

Example of plotting a streamline quiver plot.

```python
from stompy.grid import unstructured_grid
from stompy.plot import stream_quiver
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g=unstructured_grid.UnstructuredGrid()
g.add_rectilinear( p0=[0,0],p1=[5,5],nx=31,ny=31) ;
plt.figure(1)
g.plot_edges()
plt.axis('equal') ;
```

```python
cc=g.cells_center()
U=np.c_[ np.sin(cc[:,1]), np.sin(cc[:,0])]
Umag=np.sqrt( (U**2).sum(axis=1))
```

```python
zoom=[1,5,0.5,3]

# Regular quiver plot:
plt.figure()
plt.quiver( cc[:,0],cc[:,1], U[:,0], U[:,1], Umag,
           pivot='middle')
plt.axis('equal')
plt.axis(zoom) ;
```

```python
# Stream quiver:
sq=stream_quiver.StreamlineQuiver(g,U,streamline_count=200,max_dist=1.0,
                                  min_clearance=0.2)
```

```python
plt.figure()
sq.size=0.02
sq.cmap='viridis'
q=sq.plot_quiver(lw=1.2)
plt.axis('equal')
plt.axis(zoom)
```
